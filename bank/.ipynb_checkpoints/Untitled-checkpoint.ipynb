
**Below are the basic questions we will answer while exploring the dataset.**


```python
import pandas as pd
df = pd.read_csv("bank-full.csv", sep=';')
df.head()
```

**Check Dataset Info**

```python
df.info()
df.describe()
```

**Sort clients by Age**


```python
df.sort_values('age').head()
df.sort_values('age', ascending=False).head()
```

**Count customers in each job category**


```python
df['job'].value_counts()
```

**Sort job categories based on frequency**


```python
df['job'].value_counts().sort_values(ascending=False)
```

**Sort dataset by call duration (longest calls first)**


```python
df.sort_values('duration', ascending=False).head(10)
```

**Sort by number of contacts made during the campaign**


```python
df.sort_values('campaign', ascending=False).head()
```

**Which month has the highest subscription rate?**


```python
df.groupby('month')['y'].value_counts(normalize=True).unstack().fillna(0)
```

**Sort education categories by number of clients**


```python
df['education'].value_counts().sort_values(ascending=False)
```

**Top 10 clients with highest balance who subscribed**


```python
df[df['y'] == 'yes'].sort_values('balance', ascending=False).head(10)
```

**Sort by balance**

```python
df.sort_values("balance").head()
```

**Sort by marital**

```python
df.sort_values("marital").head()
```

**Bottom 10 lowest balances**

```python
df.sort_values("balance").head(10)
```

**Sort by day**

```python
df.sort_values("day").head()
```

**single marital + sorted by balance**

```python
df[df["marital"] == "single"].sort_values("balance").head()
```

**Unique job types**

```python
df['job'].unique()
```

**Duration average**

```python
df['duration'].mean()
```

**Which education group has highest subscription percentage?**

```python
df.groupby('education')['y'].value_counts(normalize=True).unstack()
```

**Find how many people have both housing AND personal loan**

```python
df[(df['housing']=='yes') & (df['loan']=='yes')].shape[0]
```

**Find the top 5 jobs with highest average balance**

```python
df.groupby('job')['balance'].mean().sort_values(ascending=False).head()
```

**Compare subscription rate for people with personal loan vs without.**

```python
df.groupby('loan')['y'].value_counts(normalize=True)
```

**What is the correlation between numeric features (age, balance, duration, campaign)?**

```python
df[['age','balance','duration','campaign']].corr()
```
